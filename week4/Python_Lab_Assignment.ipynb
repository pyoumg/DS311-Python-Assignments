### Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#  Sort and display the top five schools in overall passing rate

school_df = school_data_complete[["school_name", "budget", "size","type"]].drop_duplicates().reset_index(drop=True)
school_df=pd.DataFrame(school_df)
# Print the resulting dataframe
school_df
    

```

```python
# Calculate total school budget
school_df["budget"]
```

```python
# Calculate per student budget

school_df["per_student_budget"]=school_df["budget"]/school_df["size"]
school_df
```

```python
# Cacluate the avg math and reading score
# Find the total counts of math and reading result


avg_math= school_data_complete.groupby("school_name")["math_score"].mean()
avg_reading= school_data_complete.groupby("school_name")["reading_score"].mean()


school_df = school_df.merge(avg_math, on="school_name", how="left")
school_df = school_df.merge(avg_reading, on="school_name", how="left")


school_df
```

#### Find the passing rate for math and reading (above 70 points)

```python
# Find the counts for math result in each school that pass 70 or higher
# Calculate the math passing rate
pass_math=school_data_complete[school_data_complete["math_score"]>=70].groupby("school_name").size()
total_num=school_data_complete.groupby("school_name").size()

math_rate=pass_math/total_num*100



```

```python
# Find the counts for read result in each school that pass 70 or higher
# Calculate the read passing rate
pass_reading=school_data_complete[school_data_complete["reading_score"]>70].groupby("school_name").size()
reading_rate=pass_reading/total_num*100

school_df = school_df.merge(math_rate.rename("math_rate"), on="school_name", how="left")
school_df = school_df.merge(reading_rate.rename("reading_rate"), on="school_name", how="left")


school_df
```

```python
# Calculate the overall passing rate (average of the math and reading passing rate)

school_df["overall_rate"]=(school_df["math_rate"]+school_df["reading_rate"])/2

school_df
```

```python
#sort

school_df.sort_values(by=["overall_rate"],ascending=False).head(5)
```

```python
#  Sort and display the worst five schools in overall passing rate

school_df.sort_values(by=["overall_rate"]).head(5)
```
